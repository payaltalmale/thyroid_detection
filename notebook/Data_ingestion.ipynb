```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
```

```python
#"Reading the dataset"
df = pd.read_csv("thyroid.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#Saving the first character in a new column. Because thats what matter for this problem statement.
df["outcome"] = df["class"].str[0]
df.drop(columns="class", inplace=True)
```

```python
#Replacing all possible disease outcomes into one category - "yes".
label_list = ['S', 'F', 'A', 'R', 'I', 'M', 'N', 'G', 'K', 'L', 'Q', 'J',
       'C', 'O', 'H', 'D', 'P', 'B', 'E']
df['outcome'].replace(to_replace=label_list, value="yes", inplace=True)
```

```python
df['outcome'].replace({"-": 0, "yes": 1}, inplace=True)
```

```python
#saving the processed data 
df.to_csv("data_processed.csv", index=False)
```
